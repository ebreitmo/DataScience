```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np

# -------------------- Function computeCost ----------------------------
def computeCost(X, y, theta):
 m=len(y)
 Y=y.reshape(m,1)
 J=0.5/m*sum(np.square((np.dot(X,theta)-Y)))
 return (J)


fig = plt.figure()
ax = fig.gca(projection='3d')

x, y = np.loadtxt('ex1data1.txt', delimiter=',', unpack=True)
m = len(y)  
X=np.c_[np.ones(m),x]

print('Visualizing J(theta_0, theta_1) ...')

# Grid over which we will calculate J
theta0_vals = np.linspace(-10, 10, 100)
theta1_vals = np.linspace(-1, 4, 100)
xv, yv = np.meshgrid(theta0_vals, theta1_vals)

# Initialize J_vals to a matrix of 0's
J_vals = np.zeros((len(theta0_vals), len(theta1_vals)))

# Fill out J_vals
for i in range(0,len(theta0_vals)-1):
    for j in range(0,len(theta1_vals)-1) :
      t = [theta0_vals[i], theta1_vals[j]]
      J = computeCost(X, y, t)
      J_vals[i][j] = J[0]

# Plot the surface.
surf = ax.plot_surface(theta1_vals,theta0_vals , J_vals,cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

#Customize the z axis.
ax.set_zlim(0, 800)
#ax.zaxis.set_major_locator(LinearLocator(10))
#ax.zaxis.set_major_formatter(FormatStrFormatter('%50f'))
   
plt.show()
```

```python
fig, ax = plt.subplots()
CS = ax.contour(theta1_vals,theta0_vals , J_vals)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title('Simplest default with labels')
```
